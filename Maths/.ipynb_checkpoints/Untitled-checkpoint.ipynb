```python
# Tell Jupyter to display plots in this notebook.
%matplotlib inline

# Import the numpy package with the alias np.
import numpy as np               

# Import the pyplot package from matplotlib with the alias plt.
import matplotlib.pyplot as plt  
```

```python
sample_rate = 150.0
sample_interval = 1.0 / sample_rate
t = np.arange(0, 1, sample_interval)


x = np.linspace(0.0, 10.0 * np.pi, 100)
y = np.sin(x)
z = np.abs(np.fft.fft(y))
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(np.fft.fftshift(z))
plt.show()
```
