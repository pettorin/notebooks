```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from cosmicpy import *
```

```python
cosmo = cosmology()
```

```python
print cosmo
```

```python
k = np.logspace(-3,0,100)
plt.loglog(k,cosmo.pk(k), label='Smith et al. $P(k)$')
plt.loglog(k,cosmo.pk_lin(k), label='Linear $P(k)$')
plt.legend()
plt.xlabel('k [h/Mpc]')
plt.show()
```

```python
plt.loglog(k, cosmo.pk(k, type='eisenhu'), label='eisenhu')
plt.loglog(k, cosmo.pk(k, type='eisenhu_osc'), label='eisenhu_osc')
plt.legend()
plt.xlabel('k [h/Mpc]')
plt.title('Matter power spectrum')
plt.show()
```

```python
a = z2a(np.array([0, 1, 2, 5]))
plt.loglog(k, cosmo.pk(k,a))
plt.xlabel('k [h/Mpc]')
plt.title('Matter power spectrum at z=0, 1, 2, 5')
plt.show()
```
