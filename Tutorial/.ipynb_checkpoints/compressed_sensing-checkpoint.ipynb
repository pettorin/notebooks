## 1D Example

To get a more intuitive understanding of how the principles of CS work we begin with an example of a very simple 1D signal. 

First we define a sparse signal, $\alpha$, with 128 coefficients of which only 5 have non-zero values (*i.e.* a $k$-sparse singal with $k=5$).

```python
# full sample size
n_samples = 128

# number of peaks
n_peaks = 5

# random seed used to ensure reproducibility of results
np.random.seed(0)

# generate singal alpha
alpha = np.random.permutation(np.hstack((np.random.randn(n_peaks), np.zeros(n_samples - n_peaks))))

# plot real and imaginary components of alpha
plot.stem_plot(alpha, r'$\alpha$', imag=True)
```

Next we will generate a new signal $x$ which is the Fourier transform of $\alpha$. *i.e.*

$$x = \phi\alpha$$

where $\phi$ is a normalised FFT

$$\phi\alpha = \frac{1}{\sqrt{n}} \mathcal{F}(\alpha)$$

and $n$ is the size of the signal. Conversely,

$$\phi^{-1}\alpha = \sqrt{n} \mathcal{F}^{-1}(\alpha)$$


```python
# normalised FFT
def fftc(data):
    
    return 1 / np.sqrt(data.size) * np.fft.fftshift(np.fft.fftn(np.fft.ifftshift(data)))

# normalised inverse FFT
def ifftc(data):

    return np.sqrt(data.size) * np.fft.fftshift(np.fft.ifftn(np.fft.ifftshift(data)))

# generate signal x
x = fftc(alpha)

# plot real and imaginary components of x
plot.stem_plot(x, r'$x$', imag=True)
```

To simulate a CS problem we need to randomly subsample $x$ to generate an observation (or acquisition) $y$. In this case we will keep only a quarter of the coefficients in $x$.

We can model $y$ as 

$$y = H\phi\alpha$$

where $H$ is a masking operator.

```python
# set the number of subsamples
n_subsamples = n_samples // 4

# generate a mask
mask = np.random.permutation(np.hstack((np.ones(n_subsamples), np.zeros(n_samples - n_subsamples))))

# generate the subsampled observation y
y = sparse.mask_op(x, mask)

# plot real and imaginary components of y
plot.stem_plot(sparse.upsample(y, mask), 'y', imag=True)
```

The problem we aim to solve is to recover $\alpha$ from $y$ given that we know both $\phi$ and $H$. This corresponds to the following optimisation problem

$$\begin{aligned} & \underset{\alpha}{\text{argmin}} & \frac{1}{2}\|y-H\phi\alpha\|_2^2 + \lambda\|\alpha\|_1\end{aligned}$$

Solving this equation for a given estimate of $\alpha$ provides the cost of a given solution.

The gradient of the above problem is

$$\nabla F(\alpha) = \phi^{-1}H^T(H\phi\alpha-y)$$


Now that we have the gradient we can define a more sophisticated optimisation algorithm. For this exercise we will use the forward-backward splitting algorithm with a soft thresholding operation to promote sparse solutions. It has the following form for each iteration

$$\alpha_{n+1} = \alpha_n - \gamma * \nabla F(\alpha_n)$$
$$\alpha_{n+1} = \text{ST}_\lambda(\alpha_{n+1})$$


```python
# Function to calculate the cost.
def cost_func(y, alpha_rec, mask, lambda_):
    
    return (0.5 * np.linalg.norm(y - sparse.mask_op(ifftc(alpha_rec), mask)) ** 2 + 
            lambda_ * sparse.l1_norm(alpha_rec))

# Function to calculate the gradient
def grad(y, alpha_rec, mask):

    return ifftc(sparse.upsample(sparse.mask_op(fftc(alpha_rec), mask) - y, mask))

# Function that performs simple forward backward splitting.
def forwardBackward(observation, first_guess, mask, lambda_, n_iter=300, gamma=1.0):
    
    alpha_rec = first_guess
            
    for i in range(n_iter):
        
        alpha_temp = alpha_rec - gamma * grad(observation, alpha_rec, mask)
        alpha_rec = sparse.soft_thresh(alpha_temp, lambda_)
        cost = cost_func(observation, alpha_rec, mask, lambda_)
    
    return alpha_rec
```

Now we can use the forward-backward algorithm to recover $\alpha$. Notice how the solution improves after a given number of iterations until converges on a very good reconstruction of $\alpha$.

```python
# Interactive function to plot the recovered alpha after a given number of iterations
def solve(n_iter):
    
    alpha_rec = forwardBackward(observation=y, first_guess=np.ones(n_samples), mask=mask, 
                                lambda_=0.01, n_iter=n_iter, gamma=1.0)
    
    plot.stem_plot(alpha_rec, r'$\hat{\alpha}$', imag=True)
        
slider = IntSlider(value=1, min=1, max=201, step=40)
interact(solve, n_iter=slider)
```

Finally, we can test the quality of our solution by comparing our recovered $\hat{\alpha}$ with the true $\alpha$. Notice that the difference between the two signals is extrmeley small even for the imaginary components.

```python
alpha_rec = forwardBackward(y, np.ones(n_samples), mask, 0.01, 200, 1.0)

plot.stem_plot(alpha - alpha_rec, r'$\alpha - \hat{\alpha}$', imag=True)
```

##  2D Exercise

For this exercise you are given a subsampled noise observation $y$ and a 2D mask $H$. Your task is to recoever the original image $\hat{x}$. 

```python
# Define the image shape
image_shape = (100, 100)

# Load the mask
mask = np.load('cs_data/cs_mask.npy')

# Load the observation
y = np.load('cs_data/cs_obs_data.npy')

# Display the mask H
plot.display(mask, 'Mask', shape=image_shape, cmap='bone')

# Display the observation y
plot.display(y, r'$y$', shape=(20, 10))
```

```python
# Recover the sparse singal alpha
alpha_rec = forwardBackward(y, np.ones(mask.size), mask, 0.5, n_iter=300)

# Display the recovered signal x
plot.display(ifftc(alpha_rec), title=r'$\hat{x}$', shape=image_shape)
```
