# Compressed Sensing

```python
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt
from matplotlib import pylab

# Import interaction module
from ipywidgets.widgets import *
```

```python
pylab.rcParams['figure.figsize'] = (14.0, 8.0)
```

```python
def fftc(x):
    
    return 1 / np.sqrt(x.size) * np.fft.fftshift(np.fft.fftn(np.fft.ifftshift(x)))

def ifftc(x):

    return np.sqrt(x.size) * np.fft.fftshift(np.fft.ifftn(np.fft.ifftshift(x)))
```

```python
# Function to mask a signal.
def mask_op(signal, mask):
    
    return signal[np.where(mask == 1)[0]]

# Function to upsample a signal.
def upsample(signal, mask, dtype=complex):
    
    val = np.copy(mask).astype(dtype)
    val[val == 1] *= signal
    return val

# Function to measure the l1 norm.
def l1_norm(signal):
    
    return np.sum(np.abs(signal))
```

```python
def soft_thresh(data, threshold):
    
    return np.around(np.maximum((1.0 - threshold /
                             np.maximum(np.finfo(np.float64).eps, np.abs(data))),
                             0.0) * data, decimals=15)
```

```python
def stem_plot(data, title='Example', imag=False, mask=None):
    
    x_vals = np.arange(n_samples)
    
    if not isinstance(mask, type(None)):
        x_vals = mask_op(x_vals, mask)
    
    (markers, stemlines, baseline) = plt.stem(x_vals, np.real(data), label='Real')
    plt.setp(stemlines, linestyle="-", color="grey", linewidth=0.5 )
    plt.setp(markers, marker='o', color="#19C3F5")
    plt.setp(baseline, linestyle="--", color="grey", linewidth=2)

    if imag:

        (markers, stemlines, baseline) = plt.stem(x_vals, np.imag(data), label='Imaginary')
        plt.setp(stemlines, linestyle="-", color="grey", linewidth=0.5 )
        plt.setp(markers, marker='o', color="#EA8663")
        plt.setp(baseline, linestyle="--", color="grey", linewidth=2)
        plt.legend(loc=1)
    
    plt.xlim(0, n_samples)
    plt.title(title, fontsize=20)
    plt.show()
```

```python
n_samples = 128 # full sample size
n_peaks = 5 # number of peaks
n_subsamples = n_samples // 4 # subsample size
```

```python
np.random.seed(0)

alpha = np.random.permutation(np.hstack((np.random.randn(n_peaks), np.zeros(n_samples - n_peaks))))

stem_plot(alpha, r'$\alpha$', imag=True)
```

```python
x = fftc(alpha)

stem_plot(x, r'$x$', imag=True)
```

```python
mask = np.zeros(n_samples)
mask[np.random.permutation(n_samples)[:n_subsamples]] = 1

y = mask_op(x, mask)

stem_plot(y, 'y', imag=True, mask=mask)
```

```python
def grad(y, alpha_rec, mask):

    return ifftc(upsample(mask_op(fftc(alpha_rec), mask) - y, mask))

# Function to calculate the cost.
def cost_func(y, alpha_rec, mask, lambda_):
    
    return (0.5 * np.linalg.norm(y - mask_op(ifftc(alpha_rec), mask)) ** 2 + lambda_ * l1_norm(alpha_rec))

# Function that performs simple forward backward splitting.
def forwardBackward(observation, first_guess, mask, lambda_, n_iter=300, gamma=1.0):
    
    alpha_rec = first_guess
            
    for i in range(n_iter):
        
        alpha_temp = alpha_rec - gamma * grad(observation, alpha_rec, mask)
        alpha_rec = soft_thresh(alpha_temp, lambda_)
        cost = cost_func(observation, alpha_rec, mask, lambda_)
        
#         if not i % 20:
#             print('Iteration', i + 1, 'Cost =', cost)
    
#     return alpha_rec
    stem_plot(alpha_rec, imag=True)
```

```python
def display(data, title='example', shape=None, cmap='gist_stern'):
    
    if not isinstance(shape, type(None)):
        data = data.reshape(shape)
    
    plt.imshow(np.abs(data), cmap=cmap)
    plt.title(title, fontsize=20)
    plt.colorbar()
    plt.show()
```
